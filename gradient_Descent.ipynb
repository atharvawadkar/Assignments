```python
import numpy as np
import math
import matplotlib.pyplot as plt

class GradientDescent:
    
    def __init__(self, Iterations):
        self.__learning_rate = 0.001
        self.__precision = 0.000001
        self.__w0 = 2
        self.__gradients = list()
        self.iterations = Iterations
        
    @property
    def learning_rate(self):
        return self.__learning_rate
    @property
    def precision(self):
        return self.__precision
    @property
    def w0(self):
        return self.__w0
    @property
    def gradients(self):
        return self.__gradients
    
    def derivative(self, weight):
        answer = 2 * (weight + 3)
        return answer
    
    def loss(self):
        self.iter_list = list()

        self.curr_iter = 0
        self.previous_weight = self.__w0

        self.__gradients.append(self.previous_weight)
        self.iter_list.append(self.curr_iter)

        self.curr_weight = self.previous_weight - self.__learning_rate* (self.derivative(self.previous_weight))
        
        self.curr_iter += 1
        while( (abs(self.curr_weight - self.previous_weight) > self.__precision) and (self.curr_iter < self.iterations)):
            dydx = self.derivative(self.previous_weight)
            self.previous_weight = self.curr_weight
            self.curr_weight = self.curr_weight - self.__learning_rate * (dydx)
            self.__gradients.append(self.curr_weight)
            self.iter_list.append(self.curr_iter)
            self.curr_iter += 1
            print("Weight: {} \t\t Iterations: {} / {}".format(self.curr_weight, self.curr_iter, self.iterations))
    
    def plot(self):
        plt.figure(figsize=(10,5))
        plt.xlabel("Weights")
        plt.ylabel("Iterations")
        plt.grid()
        plt.plot(self.__gradients, self.iter_list)
        plt.show()

            


if __name__ == "__main__":
    weights = GradientDescent(Iterations=3000)
    weights.loss()
    weights.plot()

    
    
```
